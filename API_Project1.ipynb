```python
#Imports
import urllib
from PIL import Image
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np  

```

```python
#In Class Code

base_url = "https://api.census.gov/data/timeseries"
mods = "/eits/hv?"
variables = "get=cell_value,time_slot_id,category_code,seasonally_adj,\
data_type_code&geo_level_code&time=from+2000+to+2010&key="
api_url = base_url+mods+variables

api_data = urllib.request.urlopen(api_url)
data = api_data.read().decode()
js_data = json.loads(data)
census_data = pd.DataFrame(js_data)
census_data
```

```python
#Data Overview

census_data.columns = census_data.iloc[0] #Only use if column names are incorrect
census_data = census_data.drop(census_data.index[0])  #Only use if first row isn't already removed
census_data
```

```python
# A graph of home occupancy in the United States
# This was done via graphing the ratio of occupied units to total units

total_vals = census_data[census_data["data_type_code"]=="TOTAL"]
occupied_vals = census_data[census_data["data_type_code"]=="OCC"]
total_vals = total_vals[total_vals["geo_level_code"]=="US"]
occupied_vals = occupied_vals[occupied_vals["geo_level_code"]=="US"]

total_units = [int(i) for i in total_vals["cell_value"]]
occupied_units = [int(i) for i in occupied_vals["cell_value"]]

ratio = []
for i in range(len(total_units)):
    ratio.append(occupied_units[i]/total_units[i]*100)

plt.plot(ratio)
plt.xlabel("Quarter (From 2000-2010)")
plt.ylabel("Occupied Homes (%) ")
plt.title("Percentage of Home Occupancy in the United State")

#End of in class code
```

```python

# API info that we are using
# The goal is to graph the population of people who are citizens that were not born in the United States

api_url = "https://api.census.gov/data/2023/acs/acs1?get=NAME,B01001_001E,B05006_002E,B05006_047E,B05006_095E,B05006_130E&for=region:*&key="

print(api_url)


api_data = urllib.request.urlopen(api_url)
data = api_data.read().decode()
js_data = json.loads(data)
project_data = pd.DataFrame(js_data)

project_data

# B05006_002E -> Americans from Europe
# B05006_047E -> Americans from Asia
# B05006_095E -> Americans from Africa
# B05006_130E -> Americans from Oceania
```

```python
project_data.columns = project_data.iloc[0] #Only use if column names are incorrect
project_data = project_data.drop(project_data.index[0])  #Only use if first row isn't already removed
project_data= project_data.dropna()
```

```python
project_data = project_data.rename(columns={'B05006_002E': 'European_Americans'})
project_data = project_data.rename(columns={'B05006_047E': 'Asian_Americans'})
project_data = project_data.rename(columns={'B05006_095E': 'African_Americans'})
project_data = project_data.rename(columns={'B05006_130E': 'Oceanian_Americans'})

project_data
```

```python
project_data = project_data.drop("B01001_001E",axis=1)
project_data = project_data.drop("region",axis=1)
```

```python
project_data.info()
```

```python
#Changing the type from object to integer
#Below is the template used to create the code

#df['col1'] = df['col1'].astype(int)

project_data["European_Americans"] = project_data["European_Americans"].astype(int)
project_data["Asian_Americans"] = project_data["Asian_Americans"].astype(int)
project_data["African_Americans"] = project_data["African_Americans"].astype(int)
project_data["Oceanian_Americans"] = project_data["Oceanian_Americans"].astype(int)
```

```python
#A visual representation of the the population based on region and continent

import matplotlib.ticker as mticker

names = project_data["NAME"]
european_americans = project_data["European_Americans"]
asian_americans = project_data["Asian_Americans"]
african_americans = project_data["African_Americans"]
oceanian_americans = project_data["Oceanian_Americans"]


x = np.arange(len(names))  
width = 0.2 

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - 1.5 * width, european_americans, width, label="European Americans")
bars2 = ax.bar(x - 0.5 * width, asian_americans, width, label="Asian Americans")
bars3 = ax.bar(x + 0.5 * width, african_americans, width, label="African Americans")
bars4 = ax.bar(x + 1.5 * width, oceanian_americans, width, label="Oceanian Americans")

# This helps fix the the 1e6 at the top
# got this peice of code from https://discourse.matplotlib.org/t/getting-started-with-tick-locator-and-formatter-on-dates-and-logs/22522
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{int(x):,}"))

ax.set_xlabel("Region")
ax.set_ylabel("Population")
ax.set_title("Immigrant Population Based on Continent and Region")
ax.set_xticks(x)
ax.set_xticklabels(names)
ax.legend()

# I got the info on how to graph this from https://pythonguides.com/matplotlib-multiple-bar-chart/
```

# What is the immigrant population like based on contitent of origin and region?

To figure this out, I pulled columns of data from the census api for region, contitent of origin, and number of people from the continent of origin. I then put this onto a bar grph to visualize the population from each region.

Notable observations from this graph are:

1. Asian Americans are always the highest population for each region
2. Americans that originate from Oceania are always the lowest
3. The West has the highest amount of Asian Americans and Oceanian Americans
4. The Northeast Region has the highest amount of European Americans
5. The South has the most African Americans

What we can conculde from this data is that no matter what region, people who immigrated to America from Asia will always have the highest population.

Things that could cause incorrect data here are

1. There was no data for people strictly from South America and Central America so they are not represented here
2. This goes over regions as a whole and does not accuratly depict the demographic of cities or individual states
